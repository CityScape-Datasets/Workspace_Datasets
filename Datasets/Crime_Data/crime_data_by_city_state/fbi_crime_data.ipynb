## ALABAMA

```
alabama = pd.read_excel('/content/alabama.xls')
print(alabama.shape)
alabama.head()
```

```
alabama.tail(10)
```

There is only one city reported for this state by the FBI

### Data Cleaning

#### Change column names

```
header_row = 3
alabama.columns = alabama.iloc[header_row]
alabama.head()
```

```
alabama.columns = alabama.columns.str.replace("\n", " ")
```

#### Delete first 4 rows

```
alabama = alabama.iloc[4:]
```

#### Delete tail

```
alabama = alabama.drop(alabama.tail(2).index)
```

#### Add State column

```
alabama.insert(loc=1, column = 'State', value='AL')
```

#### View cleaned df

```
alabama.head()
```

### Create csv

```
alabama.to_csv('alabama.csv')
```

```
!cp alabama.csv "drive/My Drive/"
```
